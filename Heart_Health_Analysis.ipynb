```python
Avg_Age=df['age'].mean()
print(Avg_Age)
```

```python
df['sex'].value_counts()
```

### 3. What is the average resting blood pressure of patients?

```python
df['trestbps'].mean()
```

### 4. How many patients have fasting blood sugar levels higher than 120 mg/dl?

```python
df['fbs'].sum()
```

### 5. What are the different types of chest pain recorded in the dataset?

```python
df['cp'].unique()
```

### 6. What is the maximum heart rate achieved by patients?

```python
df['thalach'].max()
```

### 7. What percentage of patients experience exercise-induced angina?

```python
df['exang'].mean() * 100
```

### 8. What is the average cholesterol level in the dataset?

```python
df['chol'].mean()
```

### 9. How many patients have a resting electrocardiographic result of 2?

```python
df['restecg'].value_counts()[2]
```

### 10. What is the distribution of the number of major vessels colored by fluoroscopy?

```python
df['ca'].value_counts()
```

### 11. What is the correlation between age and cholesterol levels?

```python
df[['age', 'chol']].corr()
```

### 12. What is the distribution of chest pain types across different age groups?

```python
df.groupby('age')['cp'].value_counts().unstack().plot(kind='bar', stacked=True)
```

### 13. How does maximum heart rate vary with exercise-induced angina?

```python
df.groupby('exang')['thalach'].mean()
```

### 14. Is there a significant difference in resting blood pressure between male and female patients?

```python
df.groupby('sex')['trestbps'].mean()
```

### 15. What is the relationship between fasting blood sugar levels and the presence of heart disease?

```python
pd.crosstab(df['fbs'], df['target']).plot(kind='bar', stacked=True)
```

### 16. How does the number of major vessels (ca) affect the target variable (heart disease presence)?

```python
pd.crosstab(df['ca'], df['target']).plot(kind='bar', stacked=True)
```

### 17. What is the average oldpeak value for patients with different types of chest pain?

```python
df.groupby('cp')['oldpeak'].mean()
```

### 18. Analyze the distribution of thalassemia types (thal) among patients with heart disease.

```python
pd.crosstab(df['thal'], df['target']).plot(kind='bar', stacked=True)
```

### 19. What are the most common combinations of risk factors in patients with heart disease?

```python
df[df['target'] == 1].groupby(['cp', 'fbs', 'exang', 'thal']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```

### 20. What is the effect of combining multiple risk factors (age, cholesterol, blood pressure) on the likelihood of heart disease?

```python
sns.pairplot(df, hue='target', vars=['age', 'chol', 'trestbps'])
```

### 21. Which clinical measurement has the strongest correlation with heart disease presence?

```python
df.corr()['target'].sort_values(ascending=False)
```

### 22. How do the values of the slope of the peak exercise ST segment (slope) vary with different chest pain types?

```python
df.groupby('cp')['slope'].value_counts().unstack().plot(kind='bar', stacked=True)
```

### 23. Analyze the survival rates of patients with different thalassemia types over a period.

```python
sns.lineplot(x='thal', y='age', hue='target', data=df)
```
